```python
atheletes = pd.read_excel("/content/gdrive/My Drive/Data/Gross_Domestic_Product/Athletes.xlsx")
```

```python
athelete_count=atheletes.groupby("NOC").NOC.count().sort_values(ascending=False)
athelete_count.name = "Contingent_size"
athelete_count
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=athelete_count[:25],y=athelete_count.index[:25])
plt.title("Top 25 Countries by contingent size")
plt.xlabel("Number of atheletes")
plt.ylabel("Country")
```

```python
plt.figure(figsize=(13,4))
sns.swarmplot(y=athelete_count,size=5)
```

```python
athelete_count.describe()
```

```python
sum(athelete_count)
```

```python
contingent_size_vs_pop=pd.merge(athelete_count, population_2020, left_on ="NOC" , right_on ="Country (or dependency)" , how = "inner")
contingent_size_vs_pop
```

```python
category_of_pop=[]
for a in list(contingent_size_vs_pop["Population (2020)"]):
    if a<1000000:
        category_of_pop.append("very small")
    elif a<10000000:
        category_of_pop.append("small")
    elif a<100000000:
        category_of_pop.append("large")
    else:
        category_of_pop.append("very large")
contingent_size_vs_pop["category_of_population"] = pd.Series(category_of_pop) 
contingent_size_vs_pop
        
        
```

```python
plt.figure(figsize=(13,10))

sns.swarmplot(data=contingent_size_vs_pop,x="Contingent_size",y="category_of_population",size=4.3)
```
